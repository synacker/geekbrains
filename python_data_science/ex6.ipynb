### Задание 1

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston
```

```python
target = boston.target
target
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
X.head()
```

```python
y = pd.DataFrame(boston.target, columns=['price'])
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({'y_test' : y_test['price'], 'y_pred': y_pred.flatten()}, columns=['y_test', 'y_pred'])
check_test.head(10)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

### Заданий 4

```python
df = pd.read_csv('data/creditcard.csv')
df.describe()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
vc = df.value_counts(normalize=True)
vc.values
```

```python
df.info(verbose=True)
```

```python
pd.options.display.max_columns = 100
df.head(10)
```

```python
X = df.drop(['Class'], axis=1)
X.head()
```

```python
y = df['Class']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100, stratify=y)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
model = GridSearchCV(estimator=RandomForestClassifier(random_state=100), param_grid=parameters, scoring='roc_auc', cv=3)
model.fit(X_train, y_train)
```

```python
model.best_estimator_
```

```python
model = RandomForestClassifier(max_depth=6, max_features=3, n_estimators=15,
                       random_state=100)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict_proba(X_test)
y_pred_proba = y_pred[:, 1]
y_pred_proba
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, y_pred_proba)
```
